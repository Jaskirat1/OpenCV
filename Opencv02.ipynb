# Splitting and Merging Channels

```python
# cv2.split() --> Divides a multi-channel array into several single channel arrays

# cv2.merge() --> Merges several arrays to make a single multi-channel array. All the input matrices must have the same size.


```

```python
#Split the image into B G R components

img2 = cv2.imread('New_Zealand_Lake.jpg', cv2.IMREAD_COLOR)
img2
```

```python
b,g,r = cv2.split(img2)
```

```python
plt.figure(figsize= [20,5])
plt.subplot(1,4,1)
plt.imshow(r, cmap='gray')
plt.title("Red channel")

plt.subplot(1,4,2)
plt.imshow(g, cmap='gray')
plt.title("Green Channel")

plt.subplot(1,4,3)
plt.imshow(b, cmap = 'gray')
plt.title("Blue Channel")

Image1 = cv2.merge((b,g,r))

plt.subplot(1, 4, 4)
plt.imshow(Image1[:, :, ::-1])
plt.title("Merged Output")
```

# Converting to different Color Spaces

```python
#cv2.cvtColor() --> converts an image from one color space to another. 

'''
   The function converts an input image from one color spcae to another,
   In case of transformation to from RGB color space, the order of the channels
   should be specified explicitly(RGB or BGR).Note that the default color format 
   in OpenCV is BGR (the bytes are reversed). So the first byte in a standard(24bit)
   colot image will be an 8 bit Blue component, the second byte will be Greem , 
   and the third byte will be Red. The forth, fifth, and sixth bytes would then be
   the secod pixel(Blue, then Green , then Red), and so on.



'''


"""
Function Syntax

dst =cv2.cvtColor( src, code)

dst --> Is the output image of the same size and depth as src.


"""
```

# Changing from BGR to RGB

```python
img3 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.imshow(img3)
```

# Saving Images

```python
#cv2.imwrite() --> with two arguments.
#the first one is the filename, second argument is the image object.


#cv2.imwrite(filename, img[, params])

cv2.imwrite('New_Zealand_Lake_SAVED.png', img3)

Image(filename = 'New_Zealand_Lake_SAVED.png')
```
